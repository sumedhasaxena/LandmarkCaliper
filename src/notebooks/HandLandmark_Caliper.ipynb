# Landmark Caliper

The Landmark Caliper is designed to help stroke patients during thir rehabilitation, who have suffered loss in motor functions, especially in the movement of hand or fingers.

During rehabilitation, patients are provided with a robotic glove that they wear on the affected hand. These gloves help facilitate movement of affected hand and help patients to regain activity in the affected areas. Since each patient has unique hand geometry, these gloves needs to be of custom fit to achieve best results.

The Landmark Caliper helps in getting measurement for patients in a faster and reliable way by leveraging Machine Learning and advanced image processing techniques. The application detects location of joints in the image and produces measurements of various joints in centimeter. Measurement between two joins is the Euclidean distance. 

# Landmark Description

The model detects the keypoint localization of 21 hand-knuckle coordinates within the detected hand regions.



# Landmark Caliper Input

The Landmark Caliper takes the patient's hand image taken on a A-4 paper, with palm facing downward.

# Landmark Caliper Output

The Landmark Caliper produces following output:

1. Hand Measurements
   - This CSV file contains details of patient's hand type (Left or Right) and selected landmark distances measured in centimeters.

2. Landmark Measurements
   - This CSV file contains details of patient's hand type (Left or Right) and all landmark distances measured in centimeters (for each pair of landmarks). 

3. Landmark Image
   - Landmark drawn on the image with landmark distances
  
4. Contour Image
   - Detected A4 contours drawn on the image
